## KNN algorithm

```python
import math
def KNN(gender, age, salary, k=5):
  """
  gender: 0 means male
  gender: 1 means female
  """
  arr = []
  for i in range(len(df)):
    tempDist = math.sqrt((df.loc[i, "Gender"] - gender)**2 + (df.loc[i, "Age"] - age)**2 + (df.loc[i, "EstimatedSalary"] - salary)**2)
    arr.append([tempDist, df.loc[i, "Purchased"]])
  
  arr.sort()
  # if(arr[0][1])
  countPurchase=0
  countNotPurchase=0
  index=0
  while k:
    k -= 1
    if(arr[index][1]==0):
      countNotPurchase+=1
    else:
      countPurchase+=1
    index += 1
  print(countPurchase, countNotPurchase)
  if(countPurchase>countNotPurchase): return 1
  else: return 0
```

## Testing

```python
gender = 0
age = 35
salary = 20000

result = KNN(gender, age, salary)
if(result==1): print(f"The person will purchase the product")
else: print(f"The person will not purchase the product")
```
