## Algorithm

```python
def NaiveBayes(Gender, Age, EstimatedSalary):
  """
  returns whether the person will purchase or not?
  if purchased returns 1
  if not purchased returns 0
  """
  totalData = df.shape[0]
  print(f"Total number of data: {totalData}")

  # calculations for purchased

  result = df.loc[df['Purchased']==1]
  totalPurchased = result.shape[0]
  print(f"Total number of purchased: {totalPurchased}")

  currGenderCountPurchased = 0
  result = df.loc[df['Gender']==Gender, 'Purchased']
  for ele in result:
    if ele==1:
      currGenderCountPurchased += 1
  currGenderCountTotal = result.shape[0]
  print(f"Total number of {Gender} who purchased: {currGenderCountPurchased}")


  currAgeCountPurchased = 0
  result = df.loc[df['Age']==Age, 'Purchased']
  for ele in result:
    if ele==1:
      currAgeCountPurchased += 1
  currAgeCountTotal = result.shape[0]
  print(f"Total number of {Age} who purchased: {currAgeCountPurchased}")

  currEstimatedSalaryCountPurchased = 0
  result = df.loc[df['EstimatedSalary']==EstimatedSalary, 'Purchased'] 
  for ele in result:
    if ele==1:
      currEstimatedSalaryCountPurchased += 1
  currEstimatedSalaryCountTotal = result.shape[0]
  print(f"Total number of {EstimatedSalary} who purchased: {currEstimatedSalaryCountPurchased}")

  # handling the zero division errors
  if totalPurchased==0: totalPurchased=1
  if currGenderCountTotal==0: currGenderCountTotal=1
  if currAgeCountTotal==0: currAgeCountTotal=1
  if currEstimatedSalaryCountTotal==0: currEstimatedSalaryCountTotal=1
  if currGenderCountPurchased==0: currGenderCountPurchased=1
  if currAgeCountPurchased==0: currAgeCountPurchased=1
  if currEstimatedSalaryCountPurchased==0: currEstimatedSalaryCountPurchased=1

  finalValue1 = (totalPurchased/totalData) * (currGenderCountPurchased/currGenderCountTotal) * (currAgeCountPurchased/currAgeCountTotal) * (currEstimatedSalaryCountPurchased/currEstimatedSalaryCountTotal)

  print(f"Final non normalized value for purchased: {finalValue1}")

  # calculations for not purchased

  result = df.loc[df['Purchased']==0]
  totalPurchased = result.shape[0]
  print(f"Total number of non-purchased elements: {totalPurchased}")

  currGenderCountPurchased = 0
  result = df.loc[df['Gender']==Gender, 'Purchased']
  for ele in result:
    if ele==0:
      currGenderCountPurchased += 1
  currGenderCountTotal = result.shape[0]
  print(f"Total number of {Gender} who did not purchase: {currGenderCountPurchased}")


  currAgeCountPurchased = 0
  result = df.loc[df['Age']==Age, 'Purchased']
  for ele in result:
    if ele==0:
      currAgeCountPurchased += 1
  currAgeCountTotal = result.shape[0]
  print(f"Total number of {Gender} who did not purchase: {currGenderCountPurchased}")

  currEstimatedSalaryCountPurchased = 0
  result = df.loc[df['EstimatedSalary']==EstimatedSalary, 'Purchased'] 
  for ele in result:
    if ele==0:
      currEstimatedSalaryCountPurchased += 1
  currEstimatedSalaryCountTotal = result.shape[0]
  print(f"Total number of {EstimatedSalary} who did not purchase: {currEstimatedSalaryCountPurchased}")

  # handling the zero division errors
  if totalPurchased==0: totalPurchased=1
  if currGenderCountTotal==0: currGenderCountTotal=1
  if currAgeCountTotal==0: currAgeCountTotal=1
  if currEstimatedSalaryCountTotal==0: currEstimatedSalaryCountTotal=1
  if currGenderCountPurchased==0: currGenderCountPurchased=1
  if currAgeCountPurchased==0: currAgeCountPurchased=1
  if currEstimatedSalaryCountPurchased==0: currEstimatedSalaryCountPurchased=1

  finalValue2 = (totalPurchased/totalData) * (currGenderCountPurchased/currGenderCountTotal) * (currAgeCountPurchased/currAgeCountTotal) * (currEstimatedSalaryCountPurchased/currEstimatedSalaryCountTotal)

  print(f"Final non normalized value for non-purchased: {finalValue2}")

  normalizedFinal1 = finalValue1 / (finalValue1 + finalValue2)
  normalizedFinal2 = finalValue2 / (finalValue1 + finalValue2)

  print(f"\n\nNormalized value for purchased: {normalizedFinal1}")
  print(f"Normalized value for nonn-purchased: {normalizedFinal2}")

  if normalizedFinal1 > normalizedFinal2:
    return 1
  else:
    return 0
```

## Testing

```python
# Gender = 'Male'
# Age = 36
# EstimatedSalary = 33000

Gender = 'Female'
Age = 49
EstimatedSalary = 36000
```

```python
result = NaiveBayes(Gender, Age, EstimatedSalary)
if result==1:
  print("\n\nThe person will purchase")
else:
  print("\n\nThe person will not purchase")
```
